```python
import pandas as pd  #datayı analiz etmek ve şekillendirmek için kullanılır.
import numpy as np #bazı detaylı ve kolaylık sağlayan matematiksel işlemler için kullanılır.
```

```python
df1 = pd.read_csv("data/raw/google_patents_dataset.csv")  #ilk dataset 
df1.head() #dataseti gösterir
```

```python
df2 = pd.read_csv("data/raw/epo_patent.csv") #2.dataset
df2.head() #dataseti gösterir
```

```python
#EĞER DF2'DE PATENT_ID YOKSA O ZAMAN NAN DEĞERİ VER
df2["patent_id"] = df2.get("patent_id", np.nan) 

#TITLE SÜTUNU EKSİK DEĞERLERİ BOŞ STRING İLE DOLDUR ÇÜNKÜ BOMBOŞ
df2["title"] = df2["title"].fillna("")  

#DF2'DE ABSTRACT SÜTUNU YOKSA EKLE , YOKSA NAN İLE DOLDUR.
df2["abstract"] = df2.get("abstract", np.nan)

#df2'de assignee sütunu yoksa ekle , yoksa nan ile doldur.
df2["assignee"] = df2.get("assignee", np.nan)

#eğer df2'de inventor_author sütunu yoksa ekle , yoksa nan ile doldur.
df2["inventor_author"] = df2.get("inventor_author", np.nan)

#eğer df2'de priority_date sütunu yoksa ekle , yoksa nan ile doldur.
df2["priority_date"] = df2.get("priority_date", np.nan)
df2["filing_date"] = df2.get("filing_date", np.nan)
df2["publication_date"] = df2.get("publication_date", np.nan)
df2["grant_date"] = df2.get("grant_date", np.nan)
df2["result_link"] = df2.get("result_link", np.nan)
df2["image_link"] = df2.get("image_link", np.nan)
df2["language"] = df2.get("language", "unknown")
df2["technology_category"] = df2.get("technology_category", "other")
df2["publication_year"] = df2.get("publication_year", np.nan)
df2["filing_year"] = df2.get("filing_year", np.nan)

#df2 sütunlarını df1 ile aynı sıraya getir
df2 = df2[df1.columns]
```

```python
#concat ile dataframeleri birleştiriyoruz.
#ignore_index = True ile eski indexleri yok sayıp yeni bir index oluşturuyoruz.
merged_df = pd.concat([df1, df2], ignore_index=True)
print("Merged Shape:", merged_df.shape)
merged_df.head()
```

```python
#boyutunu yazdırıyoruz.
print("Veri Boyutu:", merged_df.shape)

#list ile dataframe ' in hangi sütunlardan oluştuğunu  öğreniyorsun
print("\nSütunlar:", list(merged_df.columns))
print("\nVeri Türleri:\n")
print(merged_df.dtypes)

# Eksik değer analizi
print("\nEksik Değer Sayıları:")
missing = merged_df.isnull().sum()
missing = missing[missing > 0]
if len(missing) > 0:
    print(missing)
else:
    print("Hiç eksik değer yok")

# Her sütunun boşluk oranı (%)
print("\nEksik Değer Yüzdeleri:")
missing_percent = merged_df.isnull().mean() * 100
print(missing_percent[missing_percent > 0].sort_values(ascending=False))
#EN COK EKSİK DEĞERDEN EN AZ OLANA GÖRE SIRALAR

#Örnek bazı satırlar
print("\nÖrnek Satırlar:")
display(merged_df.head(5))

#Benzersiz değer sayısı
print("\nHer sütundaki benzersiz değer sayısı:")
print(merged_df.nunique())
#NUNIQUE = UNİQUE OLANLARI SAYISAL OLARAK VERİR.
#UNİQUE = UNİQUE OLANLARI DİREKT GÖSTERİR.

# Dil dağılımı
print("\nDil Dağılımı:")
print(merged_df['language'].value_counts())

# Teknoloji kategorisi dağılımı
print("\nTeknoloji Kategorileri:")
print(merged_df['technology_category'].value_counts())

# Yıl sütunlarını kontrol et (float mı int mi?)
print("\nYıl Sütunları (ilk 5):")
print(merged_df[['publication_year', 'filing_year']].head())

#NaN olan image_link'leri incele
nan_images = merged_df[merged_df['image_link'].isna()] # TRUE VEYA FALSE
print(f"\nResim linki eksik satır sayısı: {len(nan_images)}") # TRUE LAR GELİR.
if len(nan_images) > 0:
    display(nan_images[['patent_id', 'title', 'language', 'technology_category']])
#DİSPLAY -> TABLO OLARAK GÖSTERİR
```

```python
merged_df.isnull().sum()
```

```python
merged_df = merged_df.dropna(subset=['patent_id','title','language'])
```

```python
merged_df.isnull().sum()
```

```python
# Küçük metin eksikleri -> 'unknown' ile doldur
merged_df.loc[:, 'inventor_author'] = merged_df['inventor_author'].fillna('unknown inventor')

# grant_date -> henüz onaylanmamış olabilir
merged_df.loc[:, 'grant_date'] = merged_df['grant_date'].fillna('not granted yet')

# Sayısal sütunlar  -> ortalama ile doldur
merged_df.loc[:, 'publication_year'] = merged_df['publication_year'].fillna(merged_df['publication_year'].mean())
merged_df.loc[:, 'filing_year'] = merged_df['filing_year'].fillna(merged_df['filing_year'].mean())
```

```python
merged_df.isnull().sum()
```

```python
merged_df['image_link'] = merged_df['image_link'].fillna('no_image')
```

```python
merged_df.isnull().sum()
```

```python
merged_df.to_csv('patentAI.csv', index=False, encoding='utf-8-sig')
#UTF-8-SİG TÜRKÇE HARFLERİ , ÇİNCE VS. DE ALABİLSİN KUSURSUZCA
print("patentAI.csv kaydedildi.")
```
